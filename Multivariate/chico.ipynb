```python
# Mobile Price Prediction using Linear Regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error
import joblib
```

```python
#  load dataset
data = pd.read_csv('train.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data.drop('Loan_ID', axis=1, inplace=True)
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data.head()
```

```python
# encode categorical data home_ownership, loan_intent, loan_grade, historical_default, Current_loan_status

encoder = LabelEncoder()
data['Gender'] = encoder.fit_transform(data['Gender'])
print(dict(zip(encoder.classes_, encoder.transform(encoder.classes_))))
data['Married'] = encoder.fit_transform(data['Married'])
print(dict(zip(encoder.classes_, encoder.transform(encoder.classes_))))
data['Education'] = encoder.fit_transform(data['Education'])
print(dict(zip(encoder.classes_, encoder.transform(encoder.classes_))))
data['Self_Employed'] = encoder.fit_transform(data['Self_Employed'])
print(dict(zip(encoder.classes_, encoder.transform(encoder.classes_))))
data['Property_Area'] = encoder.fit_transform(data['Property_Area'])
print(dict(zip(encoder.classes_, encoder.transform(encoder.classes_))))
data['Loan_Status'] = encoder.fit_transform(data['Loan_Status'])
print(dict(zip(encoder.classes_, encoder.transform(encoder.classes_))))
data['Dependents'] = encoder.fit_transform(data['Dependents'])
print(dict(zip(encoder.classes_, encoder.transform(encoder.classes_))))

```

```python
data.tail()
```

```python
x = data.drop(columns=['Loan_Status'])
y = data['Loan_Status']
```

```python
# y.head(20)
x.head(100)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
model = LinearRegression(fit_intercept=True, n_jobs=2)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
mean_abl_err = mean_absolute_error(y_test, y_pred)
print(mean_abl_err)

mean_sq_err = np.sqrt(mean_absolute_error(y_test, y_pred))
print(mean_sq_err)
# percentage score
print(model.score(x_test, y_test) * 100)
print(r2_score(y_test, y_pred) * 100)
```

```python

joblib.dump(model, 'loan_model.joblib')
```
